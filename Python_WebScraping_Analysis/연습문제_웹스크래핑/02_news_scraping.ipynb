문제 1 : 최신뉴스, 정치 , 경제, 사회, 세계, IT/과학  6개의 섹션의 뉴스를 출력하는 함수를 생성하여 스크래핑 하기

```python
import requests, os
from bs4 import BeautifulSoup
from urllib.parse import urljoin
from IPython.display import Image, display

req_header = {
    "user-agent" : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36 Edg/144.0.0.0"
}
section_dict = {
    "최신뉴스": "n0100",
    "정치": "n0200",
    "경제": "n0300",
    "사회": "n0400",
    "세계": "n0500",
    "IT/과학": "n0600",
}
def print_news(section_name):
    # 섹션 설정
    mid = section_dict.get(section_name,'n0100')
    if section_name == "최신뉴스":
        url = f"https://news.nate.com/recent?mid={mid}"
    else:
        url = f"https://news.nate.com/section?mid={mid}"
    
    print(f"======> {url} {section_name} 뉴스 <======")

    res = requests.get(url,headers=req_header)

    if res.ok:
         # Nate는 euc-kr인 경우가 많아서 이렇게 맞추는 게 안전
        res.encoding = 'euc-kr'
        soup = BeautifulSoup(res.text,'html.parser')
        # 기사 '덩어리'를 안정적으로 잡는 셀렉터
        tags = soup.select("div.postListType.noListTitle div.mlt01")
        print(type(tags),len(tags))

        for div_tag in tags:
            # 1)기사링크
            a_tag = div_tag.find("a")
            # a_tags의 href에는 http://이 부분이 없음 -> url이랑 합쳐서 사용
            link = urljoin(url, a_tag["href"])
            print(link)

            # 2)이미지
            img_tag = div_tag.select_one("span.ib img")
            if img_tag and img_tag.get("src"):
                img_url = urljoin(url,img_tag["src"])
                print(img_url)
                display(Image(url=img_url))
            
            # 3)제목
            h2_tag = div_tag.select_one("span.tb h2.tit")
            if h2_tag:
                title = h2_tag.get_text(strip=True)
                print(title)
            
            print()

    else:
        print(f"Error Code: {res.status_code}")
```
