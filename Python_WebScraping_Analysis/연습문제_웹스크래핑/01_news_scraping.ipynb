문제 1 : 경제 뉴스 URL에서 뉴스기사와 제목을 출력하기

```python
import requests
from bs4 import BeautifulSoup

url = 'https://news.daum.net/economy'
req_header = {
    'user-agent' : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36 Edg/144.0.0.0"
}
res = requests.get(url,headers=req_header)
print(url)
print(type(res))
print(res.status_code)

if res.ok:
    res.encoding = 'utf-8'
    soup = BeautifulSoup(res.text,"html.parser")
    # 클래스가 item_newsheadline2인 a태그들을 다 가져오기 [뉴스카드1, 뉴스카드2, 뉴스카드3, ...]
    a_tags = soup.select("a.item_newsheadline2")
    print(type(a_tags), len(a_tags))

    for a_tag in a_tags:
        # a태그안에서 tit_text 클래스의 텍스트 (제목) 가져오기
        title = a_tag.select_one("strong.tit_txt").get_text(strip=True)
        link = a_tag["href"]
        print(title,link)
else:
    print(f'Error code: {res.status_code}')
```

문제 2 : 여러개의 section 중 하나를 선택해서 url에서 뉴스기사의 링크와 제목을 출력하는 코드를 함수로 작성하기

```python
import requests
from bs4 import BeautifulSoup
section_dict = {
'기후/환경':'climate',
'사회':'society',
'경제':'economy',
'정치':'politics',
'국제':'world',
'문화':'culture',
'생활':'life',
'IT/과학':'tech',
'인물':'people'
}

def print_news(section_name):
    sid = section_dict.get(section_name, '사회')
    url = f"https://news.daum.net/{sid}"
    req_header = {
    'user-agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36 Edg/144.0.0.0'
    }
    
    print(f"======> {url} {section_name} 뉴스 <======")

    res = requests.get(url,headers=req_header)
    if res.ok:
        res.encoding = 'utf-8'
        soup = BeautifulSoup(res.text,'html.parser')
        a_tags = soup.select("a.item_newsheadline2")
        for a_tag in a_tags:
            title = a_tag.select_one("strong.tit_txt").get_text(strip=True)
            link = a_tag["href"]
            print(title,link)
    else:
        print(f"Error code: {res.status_code}")
    

print_news('정치')
```
